```python
from collections import Counter
```

```python
def flatten(seq, container=None):
    if container is None:
        container = []

    for s in seq:
        try:
            iter(s)  # check if it's iterable
        except TypeError:
            container.append(s)
        else:
            flatten(s, container)

    return container
```

```python
#Initialise an empty list that will be used to store the 6 lucky numbers!
lotteryNumbers = []

for i in range (0,6):
    number = random.randint(1,20)   #Check if this number has already been picked and ...
    while number in lotteryNumbers:  # ... if it has, pick a new number instead 
        number = random.randint(1,20)
    #Now that we have a unique number, let's append it to our list.
    lotteryNumbers.append(number)

lotteryNumbers.sort() #Sort the list in ascending order
print(">>> Today's lottery numbers are: ") 
print(lotteryNumbers)
```
