```python
df = df.rename(columns={'Column 1': 'Ticket #', 'Column 2': 'Winning #'})
df.head()
```

```python
df1 = df[['Ticket #', 'Winning #']]
```

```python
df1 = df1.dropna()
df1 = df1.reset_index(drop=True)
df1.head()
```

```python
win = df1['Winning #']
win = win.str.replace('|', " ")
```

```python
# Create an empty list 
Row_list =[] 
  
# Iterate over each row 
for rows in win: 
    # Create list for the current row 
    my_list = rows
    l = [int(s) for s in my_list.split() if s.isdigit()]
      
    # append the list to the final list 
    Row_list.append(l) 
  
# Print the list 
len(Row_list) 
```

```python
def flatten(seq, container=None):
    if container is None:
        container = []

    for s in seq:
        try:
            iter(s)  # check if it's iterable
        except TypeError:
            container.append(s)
        else:
            flatten(s, container)

    return container
```

```python
a = flatten(Row_list)
new = []

for i in range(1,21):
    new.append(a.count(i))
    
num = []

for i in range(1,21):
    num.append(i)
```

```python
data = {'Number':  num, 'Occurrence': new}
df2 = pd.DataFrame(data, columns = ['Number', 'Occurrence'])

df2
```

```python
df2.sort_values(by=['Occurrence'], ascending = False)
```

```python
tot = np.sum(df2["Occurrence"])

occ = df2["Occurrence"]
prob = []

for i in occ:
    temp = round(i/tot , 3)
    prob.append(temp)
    
```

```python
df2['Probability'] = prob
df2.sort_values(by=['Occurrence'], ascending = False)
```
